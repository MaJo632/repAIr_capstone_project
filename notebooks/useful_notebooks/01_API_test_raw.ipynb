```python
import requests
import json

# API endpoint URL
url = 'https://www.ifixit.com/api/2.0/guides/1220'  # Example API

# Set a limit to retrieve only 5 posts for testing purposes
params = {
    '_limit': 5  # Limits the number of posts retrieved to 5
}

# Send a GET request to the API
response = requests.get(url, params=params)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Retrieve the JSON data from the response
    data = response.json()

    # Print the data (Optional: for checking the content)
    print("Retrieved Data:")
    print(data)

    # Store the data in a local JSON file (e.g., "data.json")
    with open('data/test_API/data.json', 'w') as f:
        json.dump(data, f, indent=4)  # Save the JSON data to the file

    print("Data saved to 'data.json' successfully.")
else:
    print(f"Error: {response.status_code}")
```
